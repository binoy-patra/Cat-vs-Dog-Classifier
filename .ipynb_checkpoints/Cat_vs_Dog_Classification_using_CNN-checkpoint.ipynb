```python
#Unzip file and then create a file
import zipfile
zip_ref = zipfile.ZipFile('/content/dogs-vs-cats.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
# Importing all necessasy modules 
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
```

```python
# We can use OS Modules too for this purpose
# generators >> Devided data into batches and load batch-wise during trainning (usefull when we have large amount of data)
train_ds = keras.utils.image_dataset_from_directory(
    directory = '/content/train', #path
    labels='inferred',
    label_mode = 'int',
    batch_size=32,
    image_size=(256,256)
)

validation_ds = keras.utils.image_dataset_from_directory(
    directory = '/content/test',
    labels='inferred',
    label_mode = 'int',
    batch_size=32,
    image_size=(256,256)
)
```

```python
# Normalize : pixels value converted in the range of 0 to 1 
def process(image,label):
    image = tf.cast(image/255. ,tf.float32)
    return image,label

train_ds = train_ds.map(process)
validation_ds = validation_ds.map(process)
```

```python
# create CNN model 1

model1 = Sequential()

model1.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3))) #Convolution Layer
model1.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))                                #Pooling Layer

model1.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))                         #Convolution Layer
model1.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))                                #Pooling Layer

model1.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))                        #Convolution Layer
model1.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))                                #Pooling Layer

model1.add(Flatten())                                                                              #Flatten Layer

model1.add(Dense(128,activation='relu'))                                                           #Neural Network Layer
model1.add(Dense(64,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))                                                          #Output Layer
```

```python
model1.summary()
```

```python
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Train the model and store the result in a variable
history1 = model1.fit(train_ds,epochs=10,validation_data=validation_ds)
```

```python
# Lets visualize the accuracy graph for1 trainning and testing
import matplotlib.pyplot as plt

plt.plot(history1.history['accuracy'],color='red',label='train')
plt.plot(history1.history['val_accuracy'],color='blue',label='validation')
plt.legend()
plt.show()
```

```python
#Loss vs Val_loss graph
plt.plot(history1.history['loss'],color='red',label='train')
plt.plot(history1.history['val_loss'],color='blue',label='validation')
plt.legend()
plt.show()
```

From these above graph , it is visible that the model is overfitting. That means the model is performing well n trinning data but bad in test data.
We can prevent this by following these steps: 
1. Add more data
2. Data Augmentation -> next video
3. L1/L2 Regularizer
4. Dropout
5. Batch Normalization
6. Reduce complexity

```python
# create CNN model 2
model2 = Sequential()

model2.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3))) #Convolution Layer
model2.add(BatchNormalization())                                                                   #Hyperparameter
model2.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))                                #Pooling Layer

model2.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))                         #Convolution Layer
model2.add(BatchNormalization())                                                                   #Hyperparameter
model2.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))                                #Pooling Layer

model2.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))                        #Convolution Layer
model2.add(BatchNormalization())                                                                   #Hyperparameter
model2.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))                                #Pooling Layer

model2.add(Flatten())                                                                              #Flatten Layer

model2.add(Dense(128,activation='relu'))                                                           #Neural Network Layer
model2.add(Dropout(0.1))                                                                           #Hyperparameter
model2.add(Dense(64,activation='relu'))
model2.add(Dropout(0.1))                                                                           #Hyperparameter
model2.add(Dense(1,activation='sigmoid'))                                                          #Output Layer
```

```python
model2.summary()
```

```python
model2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history2 = model2.fit(train_ds,epochs=10,validation_data=validation_ds)
```

```python
# Lets visualize the accuracy graph for1 trainning and testing
plt.plot(history2.history['accuracy'],color='red',label='train')
plt.plot(history2.history['val_accuracy'],color='blue',label='validation')
plt.legend()
plt.show()
```

```python
#Loss vs Val_loss graph
plt.plot(history2.history['loss'],color='red',label='train')
plt.plot(history2.history['val_loss'],color='blue',label='validation')
plt.legend()
plt.show()
```
